```python
import pandas as pd
stars_w_gravity_df = pd.read_csv('https://drive.google.com/uc?id=1hf_L-8BAQTp4bSZdlMs8jQPnGpmN9Siw') #https://drive.google.com/file/d/1hf_L-8BAQTp4bSZdlMs8jQPnGpmN9Siw/view?usp=drive_link

mass_list = stars_w_gravity_df["Mass"].tolist()
radius_list = stars_w_gravity_df["Radius"].tolist()
gravity_list = stars_w_gravity_df["Gravity"].tolist()

print("Mass list:" , mass_list)
print("Radius list" , radius_list)
print("Gravity list" , gravity_list)
```

```python
mass_list_sorted = sorted(mass_list)
radius_list_sorted = sorted(radius_list)
gravity_list_sorted = sorted(gravity_list)

print("Mass list sorted:" , mass_list_sorted)
print("Radius list sorted" , radius_list_sorted)
print("Gravity list sorted" , gravity_list_sorted)
```

```python
import matplotlib.pyplot as plt

x = mass_list_sorted
y = radius_list_sorted

plt.scatter(x , y )
plt.title("Mass of Star Versus Radius")
plt.xlabel("Mass")
plt.ylabel("Radius")
plt.show()
```

```python
x = mass_list_sorted
y = gravity_list_sorted

plt.scatter(x , y )
plt.title("Mass of Star Versus Gravity")
plt.xlabel("Mass")
plt.ylabel("Gravity")
plt.show()
```

```python
x = radius_list_sorted
y = gravity_list_sorted

plt.scatter(x , y )
plt.title("Radius of Star Versus Gravity")
plt.xlabel("Radius")
plt.ylabel("Gravity")
plt.show()
```
